## **1 Estructura de control**
### **1.1 Seleccion (Sentencias condicionales)**

* Selección de una de varias alternativas en base a alguna condición.
* Indentación para estructurar el código
* Importante el símbolo ‘:’
</div>

```python
# # Asigno el bvalo uno a la variable x
x = int(input("Ingrese el valor de x"))
#Implemento una condicional para saber si x es mayor a 0
if x > 0:
    print('Valor positivo')
else:
    print('Valor no positivo')
```

<div style="font-family: 'Times New Roman', Times, serif;">

* Las condiciones son expresiones que se evalúan a True o False.
* Operadores de comparación, lógicos, de identidad y de pertenencia vistos anteriormente.
</div>

```python
x = 3
y = [1, 2, 3]
print(x > 0)
print(x > 0 and x < 10)
print(x is not y)
print(x in y)
```

<div style="font-family: 'Times New Roman', Times, serif;">

* Sepueden introducir más ‘ramas’ en sentencias condicionales a través de la palabra clave *elif*.
</div>

```python
x =-3
if x > 0:
     print('Valor positivo')
elif x == 0:
     print('Valor nulo')
else:
     print('Valor negativo')
```

<div style="font-family: 'Times New Roman', Times, serif;">

* Anidamiento.
</div>

```python
val = 120
if val > 0:
    if val < 100:
        print('Valor positivo')
    else:
        print('Valor muy positivo')
else:
    print('Valor negativo')
```

## **2 Switch**
</div>

```python
a =-6
match a:
    case _ if a > 0:
        print('positive')
    case _ if a == 0:
        print('zero')
    case _ if a < 0:
        print('negative')
```

```python
a = 'l' #week day

match a:
    case 'L':
        print("Lunes")
    case 'M':
        print("Martes")
    case 'X':
        print("Miercoles")
    case _:
        print("Wrong Day")
```

#### **Expresion ternaria**

* Estructura if-else condensado en una línea.
* Se recomienda usar solo en casos sencillos.
</div>

```python
# # # Ejemplo: cálculo del valor absoluto de un número.
x =-3
resultado = x if x >= 0 else-x
print(resultado)
```

```python
val =-3

if val >=0:
    resultado = val
else:
    resultado =-val
print(resultado)
```

#### **Comparación de variables: ‘==’ vs ‘is’**

* S‘==’ compara si el valor de las variables es el mismo
* ‘is’ compara si los objetos en las variables son iguales (referencia)
</div>

```python
a = 1000
b = a
print(a is b)
print(a == b)
c=type(a)
print(c)
d=type(b)
print(d)
print(c==d)
print(id(a))
print(id(b))
```

```python
a = [1, 2, 3]
b = [1, 2, 3]
c = a
print(a is b)
print(a is c)
print(a == b)
print(id(a))
print(id(b))
print(id(c))
```

#### **Valor booleano**

* STodos los objetos en Python tiene inherentemente un valor booleano: *True* o *False*.
* Cualquier número distinto de 0 ó cualquier objeto no vacío tienen valor *True*.
* El número 0, objetos vacíos y el objeto especial *None* tienen valor *False.*
</div>

```python
a =-10 #True
b = None #False
if a:
    print("a es True")
if not b:
    print("b es False")
```

```python
c = [1,2,3]
# a c le reasigno None
c = None #0 es igual a vacio a nada
if c:
    print("Lista no vacía")
else:
    print("Lista vacía")
```

## **3 Iteracion (bucles)**

* Repetición de un bloque de código.
* La terminación del bucle depende del tipo de bucle.
* Dos tipos: *while* y *for*.

**Bucle ‘while’**

* Repetición de un bloque de código hasta que se deje cumplir una expresión (es decir, hasta que una condición evalue a *False*).
* Si la condición evalua a False desde el principio, el bloque de código nunca se ejecuta.
* Cuidado con los bucles infinitos.

Formato general:

<p>
while test:     # Mientras se cumple la condición <br>
    atatements  # Instrucciones a ejecutar
</p>
</div>

```python
# # Ejemplo: Mostrar los primeros 3 objetos de una lista
indice = 0
numeros = [9, 4, 7, 1, 2]
while indice < 3:
    print(numeros[indice])
    indice =indice+1
```

```python
# Ejemplo: contar y mostrar los números inferiores a 10.
numeros = [33, 3, 9, 21, 1, 7, 12, 10, 8]
contador = 0
indice = 0
while indice < len(numeros):
    if numeros[indice] < 10:
        print(numeros[indice])
        contador += 1
    indice += 1
    
print('Contador:', contador)
print('Indice:', indice)
```

```python
nombre = 'Pablo'

while nombre: # Mientras 'nombre' no sea vacío
    print(nombre)
    nombre = nombre[1:]
```

```python
# #bucle infinito
i = 0
while i < 10:
    print(i)
    i += 1
```

<div style="font-family: 'Times New Roman', Times, serif;">
    
**Bucle ‘for’**

* Permite recorrer los items de una sequencia o un objeto iterable.
* Funciona en strings, listas, tuplas, etc.

Formato general:

<p>
for item in objeto: # Asigna los items del objeto a la variable item en cada iteración <br>
statements          # Instrucciones a ejecutar
</p>
</div>

```python
peliculas = ['Matrix', 'The purge', 'Avatar', 'Star Wars']

for pelicula in peliculas:
    print(pelicula)
```

<div style="font-family: 'Times New Roman', Times, serif;">

* También se usa para iterar un número preestablecido de veces (*counted loops*):
</div>

```python
for i in range(10):
    print(i)
```
