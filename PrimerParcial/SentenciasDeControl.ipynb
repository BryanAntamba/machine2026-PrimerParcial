## **3 Iteración (bucles)**

* Repetición de un bloque de código.
* La terminación del bucle depende del tipo de bucle.
* Dos tipos:*while* y *for.*

#### **Bucle ‘while’**

* Repetición de un bloque de código hasta que se deje cumplir una expresión (es decir, hasta que una condición evalue a *False*).
* Si la condición evalua a False desde el principio, el bloque de código nunca se ejecuta.
* Cuidado con los bucles infinitos.

Formato general:

<p>
    While test:   # Mientras se cumple la condición <br>
    stantements   # Instrucciones a ejecutar
</p>
</div>

```python
# # Ejemplo: Mostrar los primeros 3 objetos de una lista
indice = 0
numeros = [9, 4, 7, 1, 2]
while indice < 3:
   print(numeros[indice])
   indice =indice+1
```

```python
# Ejemplo: contar y mostrar los números inferiores a 10.
numeros = [33, 3, 9, 21, 1, 7, 12, 10, 8]
contador = 0
indice = 0
while indice < len(numeros):
   if numeros[indice] < 10:
       print(numeros[indice])
       contador += 1
   indice += 1
print('Contador:', contador)
print('Indice:', indice)
```

```python
nombre = 'Pablo'
while nombre: # Mientras 'nombre' no sea vacío
   print(nombre)
   nombre = nombre[1:]
```

```python
# #bucle infinito
i = 0
while i < 10:
    print(i)
    i += 1
```

#### **3.1 Las sentencias break, continue y else**
#### **Break y continue**

* Solo tienen sentido dentro de bucles.
* *Break* permite terminar el bucle por completo.
* *Continue* permite saltar a la siguiente iteración, continuando con el bucle.
* Pueden aparecer en cualquier parte de un bucle, pero normalmente aparecen dentro de sentencias condicionales (if).
</div>

```python
# # Ejemplo: break
rating_to_find = 4.2
movie_ratings = [4.3, 2.5, 1.7, 4.2, 3.8, 3.3, 4.5]

for rating in movie_ratings:
    print(rating)
    if rating == rating_to_find:
        print("Found")
        break
print(rating)
```

```python
for i in range(10):
    if i % 2 == 0:
        continue
# Si el número es par, salta a la siguiente iteración␣
    print(f'Odd number {i}')
```

<div style="font-family: 'Times New Roman', Times, serif;">

#### **Else**

* Los bucles pueden tener una sensencia else.
* Resulta poco intuitiva para muchos programadores porque esta sintaxis no existe en otros lenguajes.
* Se ejecuta cuando el bucle termina con normalidad; es decir, cuando no termina a causa de un break.
</div>

```python
lista = [60,60,30,60]
element_to_find = 60

for element in lista:
    if element == element_to_find:
        print("found")
        break
else:
   print("not found")
```

```python
# with flag
lista = [10,30,50,40]
element_to_find = 30
found = False

for element in lista:
    if element == element_to_find:
        found = True
        break
if found:
    print("dato encontrado")
else:
    print("not found")
```

## **4 List Comprehensions**

*  Permiten construir listas a través de la ejecución repetida (sentencia for) de una expressión para cada item de un objeto iterable.
*  Van entre ‘[’ y ’]’. Esto es indicativo de que estamos construyendo una lista.

Sintaxis:

    [<expression> for <item> in <iterable>] 
<br>  
Ejemplo: repetir los carácteres de un string.
</div>

```python
[char*3 for char in "Enrique"]
```

```python
lista = []
for char in 'Enrique':
    lista.append(char)
print(lista)
```

<div style="font-family: 'Times New Roman', Times, serif;">

Comúnmente, el item de la sentencia for aparecerá en la expresión principal, pero eso no es obli
gatorio.
</div>

```python
[2 for _ in 'Enrique']
```

<div style="font-family: 'Times New Roman', Times, serif;">

#### **4.0.1 Versión extendida**

Se puede especificar un filtro (sentencia if) para obtener únicamente los elementos que cumplan
cierta condición.
Sintaxis:

    [<expression> for <item> in <iterable> if <condition>]
<br>
Ejemplo: obtener números pares:
</div>

```python
[y for y in range(9) if y % 2 == 0]
```

<div style="font-family: 'Times New Roman', Times, serif;">

* Laslist comprehension no son realmente requeridas, ya que siempre podemos escribir un bucle equivalente.
</div>

```python
pares = []
for x in range(9):
    if x % 2 == 0:
        pares.append(x)
print(pares)
```

<div style="font-family: 'Times New Roman', Times, serif;">

#### **4.0.2 Versión completa**

La sentencia if de una list comprehension también puede contener una expresión alternativa.

Sintaxis:

    [<expression_1> if <condition> else <expression_2> for <item> in <iterable>]
<br> 
Ejemplo: poner a cero los números pares.
</div>

```python
[0 if x % 2 == 0 else x for x in range(9)]
```

```python
[x**2 if x > 2 else x for x in range(-4,5) if x > 0]
```

```python
lista = []
for x in range(-4,5):
   if x > 0:
      if x > 2:
         lista.append(x**2)
      else:
         lista.append(x)
print(lista)
```

```python
[x**2 if x > 0 else 100 if x==0 else x for x in range(-4,5)]
```

```python
lista = []

for x in range(-4,5):
    if x > 0:
        lista.append(x**2)
    elif x == 0:
        lista.append(100)
    else:
        lista.append(x)
print(lista)
```

<div style="font-family: 'Times New Roman', Times, serif;">

#### **4.0.3 Anidamiento**

Las list comprehensions soportan anidamiento en sus expresiones.
Ejemplo: bucle anidado para obtener una lista de tuplas que combinan los elementos de dos listas dadas.
</div>

```python
lista_1 = [1, 2, 3]
lista_2 = [4, 5]
[(x, y) for x in lista_1 for y in lista_2]
```

#### **4.0.4 Pros y contras**

* Principales ventajas de las comprehensions en comparación a un bucle convencional:
<ul style="font-family: 'Times New Roman', Times, serif; margin-left:40px; list-style-type:none; padding-left:0;">
    <li>- Expresión compacta y legible, si estás familiarizado con la sintaxis.</li>

#### **4.0.5 Otros tipos de comprehensions**
#### **Dictionary comprehensions**

Usando ‘{’ y ‘}’ como delimitadores y una expresión ‘clave : valor’, se obtiene un diccionario en lugar de una lista.
Ejemplo: diccionario donde cada valor es el cuadrado de la clave.
</div>

```python
d = {x : x*x for x in range(10)}
print(d)
print(type(d))
```

```python
items = ['Banana', 'Pear', 'Olives']
price = [1.1, 1.4, 2.4]
shopping = {k:v for k,v in zip(items,price)}
print(shopping)
```

```python
items = ['Banana', 'Pear', 'Olives']
price = [1.1, 1.4, 2.4]
for k in zip(items, price):
    print(type(k))
    print(k)
```

## **5 Excepciones**
* Las excepciones son eventos que representan situaciones excepcionales.
* Alteran el flujo de ejecución convencional.
* Python lanza excepciones automáticamente cuando se producen errores.
* Elprogramador puede lanzar excepciones de manera explícita y también capturar excepciones para actuar como se crea conveniente.

#### Try/except
* Permite capturar excepciones y actuar en consecuencia.

Ejemplo: error de acceso fuera de rango.
</div>

```python
lista = [6, 1, 0, 5]
lista[3]
print('Código tras el error')
```

```python
lista = [6, 1, 0, 5]
i = 300

try:
    a = lista[i]
except IndexError:
    print('He capturado la excepción de tipo IndexError')

a = 0
print('Código tras el bloque try')
print(a)
```

<div style="font-family: 'Times New Roman', Times, serif;">

* Recoger la excepción e imprimir el mensaje de error
</div>

```python
lista = [6, 1, 0, 5]
try:
   print(lista[4])
except Exception as e:
   print("Error = " + str(e))
   print('Reacheable Code')
```

<div style="font-family: 'Times New Roman', Times, serif;">

#### Try/finally
* Através de finally podemos especificar código que queremos que se ejecute siempre (indepen dientemente de si se produce la excepción o no).
* Se suele usar para liberar recursos.
</div>

```python
lista = [6, 1, 0, 5]
try:
   a = lista[1]
except IndexError:
   print('Exception IndexError Captured')
finally:
     print('Bloque finally')
     b = lista[2]
print('Código tras el bloque try')
print(b)
```

<div style="font-family: 'Times New Roman', Times, serif;">

#### raise
* La sentencia raise nos permite lanzar excepciones de manera explícita.
</div>

```python
lista = [6, 1, 0, 5]
indice = 4
try:
   if indice >= len(lista):
       raise IndexError('Índice fuera de rango')
except IndexError:
   print('Excepción de tipo IndexError capturada')
```

```python
# Se define una lista de números
listado = [1, 2, 3, 4, 5]
suma = 0
# Se recorre cada número de la lista
for numero in listado:
    # Se acumula cada valor en la variable suma
    suma += numero
print("La suma de la lista es:", suma)
```

5. Dada una lista L de números positivos, escribir un programa que muestre otra lista (ordenada) que contenga todo índice i que cumpla la siguiente condición: L[i] es múltiplo de 3. Por ejemplo, dada la lista L = [3,5,13,12,1,9] el programa mostrará la lista [0,3,5] dado que L[0], L[3] y L[5] son, respectivamente, 3, 12 y 9, que son los únicos múltiplos de 3 que hay en L.

</div>

```python
L = [3, 5, 13, 12, 1, 9]
indices = []
for i in range(len(L)):
    # Se verifica si el elemento en la posición i es múltiplo de 3
    if L[i] % 3 == 0:
        # Si cumple la condición, se agrega el índice a la lista
        indices.append(i)
indices.sort()
print("Lista de índices cuyos valores son múltiplos de 3:")
print(indices)
```

Por ejemplo, para el diccionario {‘a’: 4.3, ‘b’: 1, ‘c’: 7.8, ‘d’: -5} la respuesta sería ‘c’, dado que 7.8 es el valor más alto de los números 4.3, 1, 7.8 y -5.
</div>

```python
diccionario = {'a': 4.3, 'b': 1, 'c': 7.8, 'd': -5}
mayor_valor = None
clave_mayor = None
for clave, valor in diccionario.items():
    # Si es la primera iteración o se encuentra un valor mayor
    if mayor_valor is None or valor > mayor_valor:
        # Se actualiza el mayor valor
        mayor_valor = valor
        # Se guarda la clave correspondiente
        clave_mayor = clave
print("La clave con el valor numérico más alto es:", clave_mayor)
```
