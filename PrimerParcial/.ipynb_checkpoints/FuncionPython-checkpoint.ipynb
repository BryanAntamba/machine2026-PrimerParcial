## **Funcion Python**

Una funcion te eprmite definir un bloque de codigo reutilizable que se puede ejecutar muchas veces dentro de tu programa

Sintaxis

def nombre_funcion(vamos a recibir los parametros necesarios): bloque de codigo
</div>

```python
# funcion sin parametros o retorno de valores

def diHola():
    print("Hola programador")
```

```python
# Llamada a la funcion diHola
diHola()
```

```python
#Funcion que reciba un parametro
def holaConNombre(nombre):
    print("Hola",nombre)
```

```python
#Ingresar un nombre por teclado
nombre1=input("Ingrese el nombre del estudiante")
#Llamando a la funcion holaConNombre
holaConNombre(nombre1)
diHola()
```

```python
#Funcion que reciba dos parametros enteros por teclado y los multiplique
def multiplicacion(n1,n2):
    mul=n1*n2
    return mul
```

```python
n1=int(input("Ingrese el primer numero"))
n2=int(input("Ingrese el segundo numero"))
multiplicacion(n1,n2)
```

```python
#Crear un funcion en el cual reciba tres parametros y verifique cual de ellos es un numero mayor
def mayor(n1,n2,n3):
    if(n1>n2 and n1>n3):
        print("El numero mayor es: ",n1)
    else:
        if(n2>n1 and n2>n3):
            print("El numero mayor es: ",n2)
        else:
            print("El numero mayor es: ",n3)
```

```python
n1=int(input("Ingrese el primer numero"))
n2=int(input("Ingrese el segundo numero"))
n3=int(input("Ingrese el tercer numero"))
mayor(n1,n2,n3)
```

## **Que es un vector**

Es un espacio de memoria donde se guarda varios datos

</div>

```python
#Definimos el vector vacio
lista=[]
tamanio=int(input("Ingrese el tamaño del vector"))
for i in range(tamanio):
    n=int(input("Ingrese un valor entero"))
    lista.append(n)
#Imprima lo que tiene lista
print(lista)   
```

```python
#Crear un funcion donde se ingrese o se genere un vector con numero aleatorios del 0 al 10
import random as rd
def crearVector(n):
    vec=[]
    for i in range(n):
        num=rd.randint(0,10)
        vec.append(num)
    return vec
```

```python
n = int(input("Ingrese el tamaño del vector"))
crearVector(n)
```

```python
#Crear una funcion que reciba el tamaño de un vector e ingresar el vector por teclado
def crearVec(n):
    vec=[]
    for i in range(n):
        n=int(input("Ingrese los valores del vector"))
        vec.append(n)
    return vec
```

```python
n=int(input("Ingrese el tamaño del vector"))
crearVec(n)
```

```python
#Crear una funcion donde lea un vector y funcion donde encuentre el numero mayor
def leerVector(n):
    vector=[]
    for i in range(n):
        elemento=int(input("Ingrese el valor { i + 1 } del vector"))
        vector.append(elemento)
    return vector

def mayor(vector):
    maximo=vector[0]
    for i in vector:
        if i > maximo:
            maximo=i
    return maximo
```

```python
tamanio=int(input("Ingrese el tamaño del vector"))
vector=leerVector(tamanio)
maximo=mayor(vector)
print("Vector: ",vector)
print("El numero mayor es ",maximo)
```
