## **Normalizacion**

Ejemplo practico:

Ejercicio de Normalización

Una estación meteorológica recopila datos de 4 muestras ambientales con 4 atributos cada una: Peso (kg), Altura (cm), Edad (años) y Salario (USD). Dado que cada atributo maneja escalas completamente diferentes, se requiere aplicar una normalización min-max para transformar todos los valores al rango continuo [0, 1], de forma que los atributos sean comparables entre sí. Aplica la fórmula correspondiente por columna de forma independiente y verifica que el mínimo de cada columna sea 0 y el máximo sea 1.
Los datos registrados son los siguientes:

Muestra 1: Peso = 10 kg, Altura = 200 cm, Edad = 30 años, Salario = 5000 USD
Muestra 2: Peso = 40 kg, Altura = 150 cm, Edad = 25 años, Salario = 8000 USD
Muestra 3: Peso = 20 kg, Altura = 300 cm, Edad = 45 años, Salario = 3000 USD
Muestra 4: Peso = 50 kg, Altura = 100 cm, Edad = 20 años, Salario = 12000 USD

</div>

```python
# numpy nos da operaciones vectorizadas (min, max) por columna
# pandas nos permite visualizar la matriz como tabla legible
import numpy as np
import pandas as pd
```

```python
# Definimos la matriz de datos (4 filas = muestras, 4 columnas = atributos)
datos = np.array([
    [10, 200, 30,  5000],
    [40, 150, 25,  8000],
    [20, 300, 45,  3000],
    [50, 100, 20, 12000],
], dtype=float)

columnas = ['Peso', 'Altura', 'Edad', 'Salario']
df_original = pd.DataFrame(datos, columns=columnas)
#Visualizacion de matriz original
df_original
```

```python
# axis=0 → operación a lo largo de las filas → resultado por columna
col_min = datos.min(axis=0)
col_max = datos.max(axis=0)

print("Mínimo por columna:", col_min)
print("Máximo por columna:", col_max)
print("Rango  por columna:", col_max - col_min)
```

```python
# Fórmula: x_norm = (x - min(columna)) / (max(columna) - min(columna))
rango = col_max - col_min
datos_norm = (datos - col_min) / rango

df_norm = pd.DataFrame(datos_norm, columns=columnas)
print("Matriz normalizada [0, 1]:")
df_norm.round(4)
```

```python
# Tras la normalización el mínimo debe ser 0 y el máximo 1
print("Mín por columna (debe ser 0):", datos_norm.min(axis=0))
print("Máx por columna (debe ser 1):", datos_norm.max(axis=0))
```

```python
df_comp = pd.concat(
    [df_original, df_norm.round(4)],
    keys=['Original', 'Normalizada'],
    axis=1
)
df_comp
```

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
datos_sklearn = scaler.fit_transform(datos)

print("Resultado con sklearn (debe ser idéntico):")
print(np.round(datos_sklearn, 4))
print("\n¿Igual al cálculo manual?", np.allclose(datos_sklearn, datos_norm))
```

## **Estandarizacion**

Ejemplo practico:

Una estación de monitoreo ambiental registra 4 variables atmosféricas sobre 4 muestras: Temperatura (°C), Presión (hPa), Humedad (%) y CO₂ (ppm). Debido a que las magnitudes de cada variable son muy distintas, se requiere aplicar una estandarización z-score para que cada atributo tenga media = 0 y desviación típica = 1, permitiendo comparar las variables en una misma escala. Aplica la fórmula correspondiente por columna de forma independiente y verifica que la media de cada columna sea 0 y su desviación estándar sea 1.
Los datos registrados son los siguientes:

Muestra 1: Temperatura = 22 °C, Presión = 1013 hPa, Humedad = 55 %, CO₂ = 400 ppm
Muestra 2: Temperatura = 35 °C, Presión = 980 hPa, Humedad = 80 %, CO₂ = 650 ppm
Muestra 3: Temperatura = 18 °C, Presión = 1025 hPa, Humedad = 40 %, CO₂ = 350 ppm
Muestra 4: Temperatura = 29 °C, Presión = 998 hPa, Humedad = 65 %, CO₂ = 500 ppm
</div>

```python
# Matriz de 4 muestras × 4 atributos con escalas muy distintas
# Columnas: Temperatura (°C), Presión (hPa), Humedad (%), CO2 (ppm)
datos = np.array([
    [ 22,  1013, 55,  400],
    [ 35,   980, 80,  650],
    [ 18,  1025, 40,  350],
    [ 29,   998, 65,  500],
], dtype=float)

columnas = ['Temp (°C)', 'Presión (hPa)', 'Humedad (%)', 'CO2 (ppm)']
df_original = pd.DataFrame(datos, columns=columnas)
print("Matriz ORIGINAL:")
df_original
```

```python
# axis=0 → calcula a lo largo de las filas → un valor por columna
# ddof=0 → desviación estándar poblacional (divide entre N)
# ddof=1 → desviación estándar muestral  (divide entre N-1)
media = datos.mean(axis=0)
desv  = datos.std(axis=0, ddof=0)

print("Media por columna:             ", media)
print("Desviación típica por columna: ", desv.round(4))
```

```python
# Fórmula: x_est = (x − media(columna)) / desv(columna)
# El resultado es el z-score: cuántas desviaciones típicas
# se aleja cada valor de su media.
datos_est = (datos - media) / desv

df_est = pd.DataFrame(datos_est, columns=columnas)
print("Matriz ESTANDARIZADA (z-score):")
df_est.round(4)
```

```python
# Comprobación: tras la estandarización cada columna
# debe tener media ≈ 0 y desviación típica ≈ 1
media_est = datos_est.mean(axis=0)
desv_est  = datos_est.std(axis=0, ddof=0)

print("Media por columna  (debe ser ≈ 0):", media_est.round(10))
print("Desv.  por columna (debe ser ≈ 1):", desv_est.round(10))
```

```python
# El z-score indica cuántas desviaciones típicas se aleja
# cada valor de la media de su columna
for col in columnas:
    z_vals = df_est[col].round(4).tolist()
    print(f"{col:20} → z-scores: {z_vals}")
```

```python
# Comparación lado a lado con MultiIndex
df_comp = pd.concat(
    [df_original, df_est.round(4)],
    keys=['Original', 'Estandarizada (z)'],
    axis=1
)
df_comp
```

```python
from sklearn.preprocessing import StandardScaler

# StandardScaler aplica la misma fórmula z-score
# Ventaja: guarda los parámetros para escalar datos nuevos
scaler = StandardScaler()
datos_sklearn = scaler.fit_transform(datos)

print("Resultado con sklearn:")
print(np.round(datos_sklearn, 4))
print("\n¿Igual al cálculo manual?", np.allclose(datos_sklearn, datos_est))
```
