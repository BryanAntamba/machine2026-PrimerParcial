```python
# Creación aleatoria de datos
import numpy as np
X = np.array([0.5, 23, 0.3, 4.5, 0.3, 0.5,-28, -50, 60, -100, -10, -11, 13, 19, 1, 9])
```

```python
# Medidas de tendencia central
from scipy import stats
media = np.mean(X)
mediana = np.median(X)
moda,_ = stats.mode(X)
print('media: ', media)
print('mediana: ', mediana)
print('moda :', moda)
```

```python
# Medidas de posición
import numpy as np
import matplotlib.pyplot as plt

Q1 = np.percentile(X, 25)
Q3 = np.percentile(X, 75)
RIC = Q3 - Q1
lim_inf = Q1 - 1.5*RIC
lim_sup = Q3 + 1.5*RIC

print('RIC :', RIC)
print('limite inferior :', lim_inf)
print('limite superior :', lim_sup)
```

```python
# Dibuja el diagrama de box and whiskers y determina visualmente el número de outliers
plt.boxplot(X)
plt.grid()
plt.title("Diagrama de caja y bigotes")
plt.show()
```

```python
# Medidas de dispersión
desvest = np.std(X)
print('Desviación típica: ', desvest)
print('Varianza: ', desvest**2)
```

```python
# Medidas de distribución
asimetria = stats.skew(X)
curtosis = stats.kurtosis(X, fisher=True) # https://docs.scipy.org/doc/scipy/reference/generated/scipy.stats.kurtosis.html
print('Asimetría ', np.round(asimetria, 4))
print('Curtosis ', np.round(curtosis, 4))

if curtosis>0:
    print('Distribución Leptocúrtica')
elif curtosis<0:
    print('Distribución Platicúrtica')
else:
    print('Distribución Mesocúrtica')
```

```python
# Cargar el set de datos de Iris
from sklearn import datasets
from pprint import pprint
iris = datasets.load_iris()
print(iris)
X = iris.data
y = iris.target

# 2. Asimetría de la longitud del pétalo
petal_length = X[:, 2]

asimetria_petalo = stats.skew(petal_length)
print(f"Asimetría longitud del pétalo: {np.round(asimetria_petalo, 4)}")

if asimetria_petalo > 0:
    print("→ Asimetría positiva (cola hacia la derecha)")
elif asimetria_petalo < 0:
    print("→ Asimetría negativa (cola hacia la izquierda)")
else:
    print("→ Distribución simétrica")

# 3. Curtosis del ancho del sépalo

sepal_width = X[:, 1]

curtosis_sepalo = stats.kurtosis(sepal_width, fisher=True)
print(f"\nCurtosis ancho del sépalo: {np.round(curtosis_sepalo, 4)}")

if curtosis_sepalo > 0:
    print("→ Distribución Leptocúrtica")
elif curtosis_sepalo < 0:
    print("→ Distribución Platicúrtica")
else:
    print("→ Distribución Mesocúrtica")

# 4. Visualización de grafico

fig, axes = plt.subplots(1, 2, figsize=(12, 4))

axes[0].hist(petal_length, bins=15, color='steelblue', edgecolor='white')
axes[0].set_title(f"Longitud del pétalo\nAsimetría: {asimetria_petalo:.4f}")
axes[0].set_xlabel("cm")
axes[0].set_ylabel("Frecuencia")
axes[0].grid(True, alpha=0.3)

axes[1].hist(sepal_width, bins=15, color='seagreen', edgecolor='white')
axes[1].set_title(f"Ancho del sépalo\nCurtosis: {curtosis_sepalo:.4f}")
axes[1].set_xlabel("cm")
axes[1].set_ylabel("Frecuencia")
axes[1].grid(True, alpha=0.3)

plt.tight_layout()
plt.show()
```

```python
# Cargar datos
iris = datasets.load_iris()
X = iris.data   
y = iris.target  

nombres_clases    = iris.target_names    
nombres_variables = iris.feature_names     

# Calculo de la media y desviación típica por clase
medias = np.array([X[y == c].mean(axis=0) for c in range(3)]) 
stds   = np.array([X[y == c].std(axis=0)  for c in range(3)])  

# Representacion con plt.errorbar
fig, ax = plt.subplots(figsize=(9, 5))

colores = ['steelblue', 'seagreen', 'tomato']
x_pos   = np.arange(4)       
offset  = [-0.15, 0, 0.15]
for c in range(3):
    ax.errorbar(
        x     = x_pos + offset[c],
        y     = medias[c],
        yerr  = stds[c],
        fmt   = 'o',          
        color = colores[c],
        label = nombres_clases[c],
        capsize    = 5,       
        capthick   = 1.5,
        linewidth  = 1.5,
        markersize = 7
    )

# Ajustamiento de los ejes y etiquetas
ax.set_xticks(x_pos)
ax.set_xticklabels(nombres_variables, rotation=15, ha='right')
ax.set_ylabel('cm')
ax.set_title('Media y desviación típica por clase')
ax.legend()
ax.grid(axis='y', alpha=0.3)
plt.tight_layout()
plt.show()
```
