En las últimas décadas, el desarrollo de la inteligencia artificial ha generado una transformación profunda en la forma en que se procesan y analizan grandes volúmenes de información. Este avance ha sido impulsado principalmente por el crecimiento del aprendizaje automático y, en particular, del aprendizaje profundo, el cual ha permitido resolver problemas complejos que anteriormente resultaban difíciles de abordar mediante métodos tradicionales. Dentro de este contexto, las redes neuronales convolucionales (CNN) han emergido como una de las herramientas más relevantes, especialmente en el análisis de datos visuales, debido a su capacidad para identificar patrones y características de manera automática y eficiente.

Los antecedentes de las CNN se remontan a estudios inspirados en el funcionamiento del sistema visual humano, donde se buscaba replicar la manera en que el cerebro procesa estímulos visuales. Sin embargo, durante muchos años su desarrollo fue limitado por la falta de recursos computacionales adecuados y la escasez de grandes conjuntos de datos para su entrenamiento. Fue a partir de la década de 2010 cuando, gracias al incremento en la capacidad de procesamiento —especialmente con el uso de unidades de procesamiento gráfico (GPU) y la disponibilidad de bases de datos masivas, estas redes comenzaron a demostrar su verdadero potencial, logrando resultados sobresalientes en tareas como el reconocimiento de imágenes, la detección de objetos y el análisis de video.

Además, las CNN marcaron un cambio significativo respecto a los enfoques tradicionales, ya que eliminan en gran medida la necesidad de diseñar manualmente las características de los datos, permitiendo que el modelo aprenda directamente a partir de la información proporcionada. Esta característica no solo mejora la precisión de los resultados, sino que también optimiza los procesos de desarrollo en diversas aplicaciones tecnológicas, desde sistemas de seguridad hasta diagnósticos médicos asistidos por computadora.

En este contexto, el presente ensayo tiene como objetivo analizar las redes neuronales convolucionales, abordando sus fundamentos teóricos, su estructura y su funcionamiento, así como sus principales aplicaciones en distintos campos. ## **DESARROLLO**

Las redes neuronales convolucionales CNN, constituyen una de las arquitecturas más importantes dentro del aprendizaje profundo, debido a su capacidad para procesar información estructurada en forma de imágenes. Su funcionamiento se basa en la aplicación de filtros o kernels que recorren los datos de entrada con el fin de extraer características relevantes, tales como bordes, texturas o formas. Este proceso permite que la red aprenda de manera jerárquica, identificando desde patrones simples en las primeras capas hasta representaciones más complejas en las capas profundas. Según Yann LeCun et al. (1998), las CNN demostraron ser altamente eficaces en tareas de reconocimiento de patrones, marcando un precedente importante en el desarrollo de la visión por computadora.

En cuanto a su estructura, las CNN están compuestas principalmente por tres tipos de capas: convolucionales, de agrupamiento pooling y totalmente conectadas. Las capas convolucionales son responsables de la extracción de características mediante la aplicación de filtros, mientras que las capas de pooling reducen la dimensionalidad de los datos, lo que contribuye a disminuir el costo computacional y a evitar el sobreajuste. Finalmente, las capas totalmente conectadas se encargan de realizar la clasificación a partir de las características previamente obtenidas. De acuerdo con Ian Goodfellow et al. (2016), esta arquitectura ha permitido mejorar significativamente el rendimiento de los modelos en tareas complejas, consolidándose como una de las bases del aprendizaje profundo moderno.

Por otro lado, una de las principales ventajas de las CNN es su capacidad para aprender automáticamente las características relevantes de los datos, lo que elimina la necesidad de realizar ingeniería de características manual. Esto representa un avance importante frente a los métodos tradicionales, donde era necesario diseñar previamente los atributos que el modelo debía analizar. Sin embargo, este tipo de redes también presenta limitaciones, entre las que destacan su alto requerimiento de recursos computacionales y la necesidad de grandes volúmenes de datos para su entrenamiento. En este sentido, Geoffrey Hinton (2012) señala que el éxito de los modelos de aprendizaje profundo está estrechamente relacionado con la disponibilidad de datos y la capacidad de procesamiento.

Asimismo, las aplicaciones de las CNN son amplias y abarcan diversos campos. En la visión por computadora, se utilizan para el reconocimiento facial, la detección de objetos y la segmentación de imágenes. En el ámbito de la medicina, han demostrado ser útiles en el análisis de imágenes médicas para la detección temprana de enfermedades. También se aplican en sistemas de conducción autónoma, seguridad, agricultura de precisión y análisis de video. Estas aplicaciones evidencian la versatilidad y el impacto de las CNN en la vida cotidiana y en el desarrollo tecnológico actual.

En el mundo de las redes neuronales convolucionales no existe un único modelo, sino múltiples arquitecturas diseñadas para resolver problemas específicos con mayor eficiencia y precisión. En este apartado, se abordarán los diferentes tipos de redes neuronales convolucionales, analizando sus características principales, sus aportes al campo del aprendizaje profundo y las razones por las cuales han sido propuestas a lo largo del tiempo. Esta revisión permitirá comprender cómo la evolución de estas arquitecturas ha contribuido al avance de la inteligencia artificial, optimizando el rendimiento en tareas como la clasificación de imágenes, la detección de objetos y el procesamiento de video.

**ALEXNET** <br>

La red neuronal convolucional AlexNet representa un hito fundamental en el desarrollo del aprendizaje profundo, ya que marcó el resurgimiento de las redes neuronales en el campo de la visión por computadora. Fue propuesta en 2012 por Alex Krizhevsky, junto con Ilya Sutskever y Geoffrey Hinton, y alcanzó reconocimiento mundial tras ganar la competencia ImageNet Large Scale Visual Recognition Challenge 2012, superando ampliamente a otros métodos tradicionales.

AlexNet se caracteriza por ser una arquitectura profunda compuesta por ocho capas principales: cinco capas convolucionales y tres capas totalmente conectadas. Una de sus principales innovaciones fue el uso de la función de activación ReLU (Rectified Linear Unit), que permitió acelerar significativamente el entrenamiento en comparación con funciones tradicionales como la sigmoide. Además, incorporó técnicas como el dropout para reducir el sobreajuste y el uso de data augmentation para mejorar la generalización del modelo.

<div style="text-align: center;">
   <img src="img/arquitecturaalexnet.png" style="width:700px">
</div>


Otro aspecto clave de AlexNet fue el aprovechamiento de unidades de procesamiento gráfico (GPU), lo que permitió entrenar modelos más grandes y complejos en menos tiempo. Esta implementación demostró que el uso de hardware especializado era esencial para el desarrollo de redes neuronales profundas eficientes.

En términos de impacto, AlexNet no solo mejoró significativamente la precisión en tareas de clasificación de imágenes, sino que también impulsó una nueva era en la investigación de inteligencia artificial, motivando el desarrollo de arquitecturas más avanzadas como VGGNet, ResNet y DenseNet. En consecuencia, AlexNet es considerada una de las bases del éxito moderno del aprendizaje profundo y un punto de inflexión en la evolución de las redes neuronales convolucionales.

**VGGNET** <br>

VGGNet es una red neuronal convolucional propuesta en 2014 por el grupo de investigación Visual Geometry Group, liderado por Karen Simonyan y Andrew Zisserman. Esta arquitectura alcanzó gran reconocimiento tras su participación en la competencia ImageNet Large Scale Visual Recognition Challenge 2014, donde obtuvo resultados sobresalientes en tareas de clasificación de imágenes.

La principal característica de VGGNet es su enfoque en la simplicidad y uniformidad de la arquitectura. A diferencia de modelos anteriores que utilizaban filtros grandes, VGGNet emplea exclusivamente filtros pequeños de 3 × 3 en todas sus capas convolucionales. Esta decisión permitió aumentar la profundidad de la red sin perder capacidad de aprendizaje, ya que múltiples filtros pequeños consecutivos pueden capturar patrones complejos de manera más eficiente que un solo filtro grande.

Existen varias versiones de VGGNet, siendo las más conocidas VGG-16 y VGG-19, que hacen referencia al número de capas con pesos entrenables. En VGG-16, la red está compuesta por 13 capas convolucionales y 3 capas totalmente conectadas. La arquitectura también incluye capas de max pooling entre bloques convolucionales para reducir la dimensionalidad y mantener las características más importantes.

Una de sus ventajas más importantes es su alta precisión en tareas de clasificación de imágenes; sin embargo, también presenta una gran desventaja: su elevado costo computacional y su alto número de parámetros, lo que requiere mucha memoria y tiempo de entrenamiento. Debido a esto, aunque es muy precisa, no es la más eficiente para sistemas con recursos limitados.

**GOOGLENET** <br>
GoogLeNet abordó los desafíos de arquitecturas anteriores de CNN introduciendo el concepto de módulos iniciales. Los módulos de inicio son un tipo de componente básico que permite el procesamiento paralelo de datos a múltiples escalas. Esto permite que la red capture funciones a diferentes escalas de manera más eficiente que las arquitecturas anteriores.

Un módulo inicial normalmente consta de varias capas convolucionales con diferentes tamaños de filtro. Estas capas están dispuestas en paralelo, de modo que la red pueda procesar los datos de entrada en múltiples resoluciones simultáneamente. Luego, la salida de las capas convolucionales se concatena y se pasa a través de una capa de agrupación. Sin embargo, más tarde hubo varias versiones del módulo inicial que se integraron en consecuencia en la arquitectura que constaba de diferentes capas y patrones de tamaño de filtro.

<div style="text-align: center;">
   <img src="img/diagramagooglenet.png" style="width:700px">
</div>

Este enfoque de procesamiento paralelo tiene varias ventajas. En primer lugar, permite que la red capture funciones a diferentes escalas de manera más eficiente. Esto se debe a que la red puede procesar los datos de entrada en múltiples resoluciones simultáneamente, lo que le permite capturar características tanto a gran como a pequeña escala. En segundo lugar, ayuda a aliviar el problema de la desaparición de los gradientes. Esto se debe a que el enfoque de procesamiento paralelo permite a la red aprender funciones a múltiples escalas, lo que puede ayudar a estabilizar el proceso de entrenamiento.
<br><br>
**Convoluție 1x1:** Esta ruta aplica una convolución 1x1 a la entrada. Esto reduce la cantidad de canales en la entrada, lo que ayuda a reducir la complejidad computacional de la red.<br>
**Convoluție 3x3:** Esta ruta aplica una convolución 3x3 a la entrada. Esta es una operación convolucional estándar que se utiliza para extraer características de la imagen de entrada.<br>
**Convoluție 5x5:** Esta ruta aplica una convolución de 5x5 a la entrada. Esta ruta se utiliza para capturar características a mayor escala de la imagen de entrada.<br>
**Agrupación máxima:** Esta ruta aplica una operación de agrupación máxima a la entrada. Esta operación reduce el tamaño de la entrada manteniendo el valor máximo en cada ventana 2x2.<br><br>
Luego, las salidas de las cuatro rutas se concatenan y se aplica una convolución 1x1 al resultado. Esta convolución final 1x1 ayuda a reducir la cantidad de canales en la salida y también ayuda a mejorar la precisión de la red.

**RESNET** <br>

ResNet es una arquitectura de red neuronal convolucional propuesta en 2015 por Kaiming He, junto con Xiangyu Zhang, Shaoqing Ren y Jian Sun. Esta red destacó al ganar la competencia ImageNet Large Scale Visual Recognition Challenge 2015, logrando un avance significativo en precisión y profundidad de las redes neuronales.

La principal innovación de ResNet es la introducción de las conexiones residuales o skip connections. Estas conexiones permiten que la información de una capa anterior se transfiera directamente a capas más profundas, saltándose una o más capas intermedias. Esto se logra mediante bloques residuales, donde la red no aprende directamente una función completa, sino una función residual, es decir, la diferencia entre la entrada y la salida deseada.

Gracias a este enfoque, ResNet resolvió uno de los principales problemas de las redes profundas: el desvanecimiento del gradiente, que dificultaba el entrenamiento de modelos muy grandes. Con las conexiones residuales, fue posible entrenar redes con más de 50, 101 e incluso 152 capas, manteniendo un rendimiento estable y mejorando la precisión.

En términos de arquitectura, ResNet está compuesta por múltiples bloques residuales, cada uno formado por capas convolucionales, normalización y activación ReLU, junto con la conexión directa de entrada-salida. Además, utiliza batch normalization para estabilizar el entrenamiento y mejorar la convergencia del modelo.

Otra ventaja importante es que ResNet mejora significativamente la capacidad de generalización del modelo sin incrementar excesivamente la complejidad del entrenamiento en comparación con redes profundas anteriores como VGGNet. Esto la convierte en una de las arquitecturas más utilizadas en tareas modernas de visión por computadora.

**XCEPTION** <br>

Xception es una forma abreviada de Extreme Inception, que es profunda red neuronal convolucional (CNN) arquitectura diseñada para mejorar el original Modelo de inicio. Reemplaza las capas de convolución estándar con convoluciones separables en profundidad, lo que las hace más eficientes y precisas para tareas como la clasificación de imágenes. Esta modificación reduce el costo computacional sin comprometer el rendimiento, lo que convierte a Xception en una opción popular en las aplicaciones modernas de visión por computadora.

<div style="text-align: center;">
   <img src="img/arquitecturaxcepcion.png" style="width:450px">
</div>

La principal innovación de Xception es el uso de convoluciones separables en profundidad (depthwise separable convolutions). Este tipo de convolución divide el proceso en dos etapas: primero se aplican filtros independientes a cada canal de la imagen (convolución espacial), y luego se combinan los resultados mediante una convolución 1×1. Esto permite reducir significativamente el número de parámetros y el costo computacional, manteniendo o incluso mejorando la precisión del modelo.

En comparación con Inception, Xception simplifica la arquitectura al eliminar la necesidad de múltiples ramas paralelas, reemplazándolas por una estructura más eficiente y lineal. A pesar de esta simplificación, el modelo logra un rendimiento muy competitivo en tareas de clasificación de imágenes, especialmente en el conjunto de datos ImageNet Large Scale Visual Recognition Challenge.

Otra ventaja importante de Xception es su eficiencia, ya que reduce considerablemente la cantidad de operaciones necesarias sin sacrificar capacidad de aprendizaje. Esto lo convierte en un modelo ideal para aplicaciones donde los recursos computacionales son limitados.

**DENSENET** <br>
DenseNet es una arquitectura de red neuronal convolucional propuesta en 2017 por Gao Huang, junto con Zhuang Liu, Laurens van der Maaten y Kilian Q. Weinberger. Esta arquitectura fue presentada como una mejora en la eficiencia del flujo de información dentro de redes profundas, destacándose en competencias de visión por computadora como ImageNet Large Scale Visual Recognition Challenge.

La principal innovación de DenseNet es su estrategia de conexiones densas, donde cada capa está conectada directamente con todas las capas anteriores dentro de un mismo bloque. Esto significa que, en lugar de pasar la información solo a la siguiente capa, cada capa recibe como entrada la concatenación de todas las características previas. Esta estructura mejora significativamente la propagación de la información y del gradiente durante el entrenamiento.

Gracias a este diseño, DenseNet reduce el problema del desvanecimiento del gradiente y fomenta la reutilización de características, lo que permite construir redes más compactas y eficientes en comparación con arquitecturas anteriores como VGGNet o incluso ResNet en ciertos casos. Además, al reutilizar información ya aprendida, se reduce la cantidad de parámetros necesarios, lo que hace al modelo más eficiente en términos de memoria.

En términos de arquitectura, DenseNet está organizado en bloques densos (dense blocks) separados por capas de transición, que incluyen operaciones de convolución y pooling para reducir la dimensionalidad. Esta estructura permite mantener un flujo de información constante y bien distribuido a lo largo de la red.

**MOBILENET** <br>

MobileNet es una arquitectura de red neuronal convolucional desarrollada por Google con el objetivo de crear modelos eficientes y ligeros para dispositivos con recursos limitados, como teléfonos móviles y sistemas embebidos. Fue diseñada para aplicaciones de visión por computadora donde se requiere un equilibrio entre precisión y bajo consumo computacional.

La principal innovación de MobileNet es el uso de convoluciones separables en profundidad (depthwise separable convolutions), una técnica que divide la convolución tradicional en dos pasos: primero se aplica una convolución por cada canal de la imagen (convolución en profundidad) y luego una convolución 1×1 para combinar la información. Esta estrategia reduce significativamente la cantidad de parámetros y operaciones necesarias, haciendo que el modelo sea mucho más eficiente.

MobileNet también introduce un parámetro llamado width multiplier, que permite ajustar el número de canales en cada capa para reducir aún más el tamaño del modelo, y un resolution multiplier, que disminuye el tamaño de la imagen de entrada para acelerar el procesamiento. Estas características hacen que la red sea altamente flexible según las necesidades de la aplicación.

Las redes neuronales convolucionales (CNN) utilizan numerosos filtros predefinidos y almacenados, y realizan una convolución (X) de cada filtro con la matriz de píxeles de la imagen. Esto permite filtrar los objetos de la imagen y compararlos con un amplio conjunto de objetos predefinidos para identificar coincidencias. De esta forma, estos modelos son capaces de predecir la imagen.

Las versiones posteriores, como MobileNetV2 y MobileNetV3, mejoraron el rendimiento incorporando bloques residuales invertidos y técnicas de optimización automática de arquitectura. Estas mejoras permitieron aumentar la precisión sin comprometer la eficiencia, consolidando a MobileNet como una de las arquitecturas más utilizadas en dispositivos móviles.

**SQUEEZENET**<br>

SqueezeNet es un algoritmo diseñado para ser pequeño pero muy preciso. En 2016, investigadores de DeepScale (una nueva empresa adquirida posteriormente por Tesla), UC Berkeley y la Universidad de Stanford propusieron esta red neuronal convolucional (CNN) compacta y eficiente. El objetivo principal de SqueezeNet es lograr un equilibrio entre alta precisión y baja complejidad, lo que lo convierte en una opción ideal para dispositivos con recursos limitados como móviles y sistemas integrados.

Cada fire module está compuesto por dos partes: una capa de “squeeze”, que utiliza filtros 1×1 para reducir la cantidad de canales de entrada, y una capa de “expand”, que aplica filtros 1×1 y 3×3 para recuperar y procesar la información. Esta combinación permite disminuir drásticamente el número de parámetros sin afectar significativamente la capacidad de aprendizaje del modelo.

SqueezeNet v1.1 mejoró la precisión mediante 1,4% en el conjunto de datos de ImageNet, mientras que SqueezeNet v2.0 mejoró la precisión 1,8% con un Reducción de 2,4x en el número de parámetros.

**NASNET** <br>

NASNet es una arquitectura de red neuronal convolucional desarrollada por investigadores de Google en 2017, bajo la dirección de Barret Zoph y Quoc V. Le. Esta propuesta introdujo un enfoque innovador en el diseño de redes neuronales, al utilizar búsqueda automática de arquitecturas (Neural Architecture Search, NAS) en lugar de diseñarlas manualmente.

La idea principal de NASNet es que un algoritmo de aprendizaje por refuerzo se encarga de construir automáticamente los bloques óptimos de una red neuronal convolucional. En lugar de que un humano defina la estructura completa de la red, el sistema aprende a generar “celdas” o módulos eficientes que luego se repiten para formar la arquitectura completa.

Estas celdas se dividen en dos tipos principales: celdas normales, que mantienen la resolución de la imagen, y celdas de reducción, que disminuyen la dimensionalidad para hacer el procesamiento más eficiente. La combinación de estas celdas permite construir redes profundas y altamente optimizadas.

Una de las principales ventajas de NASNet es que las arquitecturas generadas automáticamente suelen superar en rendimiento a muchas redes diseñadas manualmente, como Inception o ResNet, especialmente en tareas de clasificación de imágenes en el conjunto de datos ImageNet Large Scale Visual Recognition Challenge. Sin embargo, su proceso de búsqueda puede ser extremadamente costoso en términos computacionales.

**U-NET** <br>

U-Net es un tipo de red neuronal que se utiliza principalmente para la segmentación de imágenes, es decir, para dividir una imagen en diferentes partes e identificar objetos específicos; por ejemplo, para separar un tumor del tejido sano en una tomografía médica. El nombre "U-Net" proviene de la forma de su arquitectura, que se asemeja a la letra "U" al dibujarla. Se utiliza ampliamente en imágenes médicas porque ofrece un buen rendimiento incluso con una pequeña cantidad de datos etiquetados

   <img src="img/arquitecturaunet.png" style="width:550px">
</div>

**Ruta de contracción** <br>
* Utiliza filtros pequeños (3×3 píxeles) para escanear la imagen y detectar características.
* Aplicar una función de activación llamada ReLU para añadir no linealidad ayuda al modelo a aprender mejor.
* Utiliza agrupamiento máximo (filtros de 2×2) para reducir el tamaño de la imagen conservando la información importante. Esto ayuda a la red a centrarse en las características más relevantes.<br><br>

**Cuello de botella:** En el centro de la "U" se almacena la información más comprimida y abstracta. Conecta el codificador con el decodificador.<br><br>
  
**Ruta expansiva** 
* Utiliza el sobremuestreo, es decir, aumenta el tamaño de la imagen para recuperar el tamaño original.
* Combina la información del codificador mediante "conexiones de salto". Estas conexiones ayudan al decodificador a obtener detalles espaciales que podrían haberse perdido al reducir el tamaño de la imagen.
* Vuelve a utilizar capas de convolución para limpiar y refinar la salida.<br><br>

La imagen superior muestra cómo U-Net transforma una imagen de 572×572 píxeles en un mapa segmentado más pequeño de 388×388 píxeles. Reduce la imagen para capturar características y luego la amplía para restaurar su tamaño original, utilizando conexiones de salto para conservar los detalles. La salida etiqueta cada píxel como objeto o el fondo. Tras comprender la arquitectura, es importante ver cómo U-Net procesa realmente los datos para realizar la segmentación:

**Imagen de entrada:** El proceso comienza introduciendo en la red una imagen médica o de otro tipo, normalmente en escala de grises.<br>
**Extracción de características:** El codificador extrae características cada vez más abstractas mediante la aplicación de convoluciones y submuestreo. En cada nivel, el tamaño espacial disminuye mientras que el número de canales de características aumenta, lo que permite al modelo capturar patrones de nivel superior.<br>
**Procesamiento de cuello de botella:** Esta es la parte central de la red donde la imagen se reduce al máximo. Contiene una versión pequeña pero muy significativa de la imagen que captura las características principales.<br>
**Reconstrucción y localización:** El decodificador comienza a reconstruir el tamaño de la imagen original mediante sobremuestreo. En cada nivel, combina las características del decodificador con las características correspondientes del codificador utilizando conexiones de salto para conservar los detalles espaciales finos.<br>
**Conexiones de salto para mayor precisión:** Las conexiones de salto ayudan a preservar la precisión espacial al incorporar características detalladas de capas anteriores. Son especialmente útiles cuando el modelo necesita distinguir límites en tareas de segmentación.<br>
**Predicción final:** Una convolución de 1×1 al final convierte los mapas de características refinados en el mapa de segmentación final, donde cada píxel se clasifica en una clase específica, como primer plano o fondo. Esta salida tiene la misma resolución espacial que la imagen de entrada.

**YOLO**

YOLO es un modelo de detección de objetos que redefine la visión por ordenador al permitir la detección en tiempo real de múltiples objetos en imágenes y vídeos, lo que lo hace bastante único respecto a los métodos de detección anteriores. En lugar de analizar cada objeto detectado individualmente, la arquitectura deYOLO trata la detección de objetos como una única tarea, prediciendo tanto la ubicación como la clase de los objetos de una sola vez mediante CNN. 

El modelo funciona dividiendo una imagen en una cuadrícula, y cada parte se encarga de detectar objetos en su área respectiva. Realiza múltiples predicciones para cada sección y filtra los resultados menos fiables, quedándose solo con los precisos.

La introducción de YOLO en las aplicaciones de visión por ordenador hizo que la detección de objetos fuera mucho más rápida y eficaz que los modelos anteriores. Gracias a su velocidad y precisión, YOLO se convirtió rápidamente en una opción popular para soluciones en tiempo real en sectores como la fabricación, la sanidad y la robótica.

Las versiones posteriores, como YOLOv2, YOLOv3 y YOLOv4, han mejorado significativamente la precisión y estabilidad del modelo, incorporando técnicas como batch normalization, anchor boxes y mejoras en la arquitectura convolucional. Estas mejoras han permitido que YOLO se mantenga como una de las arquitecturas más utilizadas en detección de objetos.

Este ejercicio consiste en utilizar una red neuronal convolucional preentrenada (MobileNetV2) para analizar una imagen llamada azul.png y determinar si la imagen contiene predominancia del color azul.

La CNN no se entrena desde cero, sino que se usa como extractor de características visuales, obteniendo una representación numérica de la imagen. Luego, esas características se usan para decidir si la imagen pertenece a la categoría “azul” o “no azul”.

```python
# Importa TensorFlow para trabajar con redes neuronales.
# Carga MobileNetV2, una CNN ligera y eficiente.
# preprocess_input prepara la imagen para la red.
# image permite cargar imágenes.
# numpy permite manejar datos numéricos.
import tensorflow as tf
from tensorflow.keras.applications import MobileNetV2
from tensorflow.keras.applications.mobilenet_v2 import preprocess_input
from tensorflow.keras.preprocessing import image
import numpy as np
```

```python
# Cargamos el modelo (MobileNetV2)
model = MobileNetV2(weights='imagenet', include_top=False, pooling='avg')
```

```python
# Convertir la imagen en numéricos, en una matriz, agregando dimensión extra para batch, y normalizando.
x = image.img_to_array(img)
x = np.expand_dims(x, axis=0)
x = preprocess_input(x)
```

```python
# Convertir la imagen en numéricos, en una matriz, agregando dimensión extra para batch, y normalizando.
x = image.img_to_array(img)
x = np.expand_dims(x, axis=0)
x = preprocess_input(x)

# Extraer características con la CNN
features = model.predict(x)

#Clasificaciómos y verificamos si es azul el color o no lo es.
threshold = np.mean(features)

if threshold > 0:
    print("La imagen probablemente contiene color AZUL")
else:
    print("La imagen NO parece ser azul")
```
